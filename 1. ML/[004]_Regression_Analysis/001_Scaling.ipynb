# **I. 실습 데이터**

> ## **1) seaborn 'mpg' Data Set**

```python
import seaborn as sns

DF = sns.load_dataset('mpg')
```

- 자동차 연비(mpg) 데이터
  - mpg : miles per galln
  - 1 mile : 1.6 Km
  - 1 gallon : 3.78 Liter

```python
DF.info()
```

```python
DF.head()
```

- X, y Data

```python
X = DF[['weight']]
y = DF['mpg']
```

> ## **2) Without Scaling**

- X값 확인

```python
X[:5]
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(9, 6))
sns.regplot(x=X, y=y)
plt.xlabel('weight_Without_Scaling')
plt.show()
```

> ## **3) With Normalization**

- sklearn Package

```python
from sklearn.preprocessing import MinMaxScaler

scaler1 = MinMaxScaler()
X_Norm = scaler1.fit_transform(X)
```

- 정규화된 X값 확인

```python
X_Norm[:5]
```

- 정규화된 X값 모델 생성

```python
fig = plt.figure(figsize=(9, 6))
sns.regplot(x=X_Norm, y=y)
plt.xlabel('weight_With_Normarlization')
plt.show()
```

> ## **4) With Standardization**

- sklearn Package

```python
from sklearn.preprocessing import StandardScaler

scaler2 = StandardScaler()
X_Stan = scaler2.fit_transform(X)
```

- 표준화된 X값 확인

```python
X_Stan[:5]
```

- 표준화된 X값 모델 생성

```python
fig = plt.figure(figsize=(9, 6))
sns.regplot(x=X_Stan, y=y)
plt.xlabel('weight_With_Standardization')
plt.show()
```
